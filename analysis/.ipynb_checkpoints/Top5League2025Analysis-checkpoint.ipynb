```python
def categorize_position(pos):
    if isinstance(pos, str):
        pos = pos.upper()
        if any(x in pos for x in ["FW", "FWD", "ST", "WING", "W"]):
            return "Attacker"
        elif any(x in pos for x in ["MF", "MID", "CM", "AM", "DM"]):
            return "Midfielder"
        elif any(x in pos for x in ["DF", "DEF", "CB", "FB", "LB", "RB"]):
            return "Defender"
    return None

df["Position_Group"] = df["Position"].apply(categorize_position)

print("Position groups assigned.")
df[["Player", "Club", "Position", "Minutes Played", "Position_Group"]].head()
```

```python
per90_cols = [
    'Goals/90', 'Assists/90', '(Goals+Assists)/90',
    'Non-Penalty Goals/90', '(Non-Penalty Goals + Assists)/90',
    'Expected Goals/90', 'Expected Assists/90',
    '(Expected Goals + Expected Assists)/90',
    'Non-Penalty Expected Goals/90',
    '(Non-Penalty Expected Goals + Expected Assists)/90',
    'Progressive Carries', 'Progressive Passes', 'Progressive Passes Received'
]

cluster_features = ['Player', 'Club', 'League', 'Group'] + per90_cols
df_cluster = df[cluster_features].copy()

df_cluster.head()
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()


scaled_data = scaler.fit_transform(df_cluster[per90_cols])


scaled_df = pd.DataFrame(scaled_data, columns=per90_cols)
scaled_df[['Player', 'Club', 'League', 'Group']] = df_cluster[['Player', 'Club', 'League', 'Group']].values

scaled_df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# To store results
clustered_dfs = {}


for group_name, group_df in scaled_df.groupby('Group'):
    print(f"--- {group_name} ---")
    X = group_df[per90_cols].values
    
    # Elbow Method
    inertia = []
    K = range(2, 10)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)

    # Plot Elbow
    plt.figure()
    plt.plot(K, inertia, 'bo-')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Inertia')
    plt.title(f'Elbow Method for {group_name}')
    plt.show()
```

```python
best_k = {
    'Attacker': 4,
    'Midfielder': 4,
    'Defender': 5
}
```

```python
from sklearn.cluster import KMeans

final_clusters = []

for group_name, group_df in scaled_df.groupby('Group'):
    X = group_df[per90_cols].values
    kmeans = KMeans(n_clusters=best_k[group_name], random_state=42, n_init=10)
    group_df = group_df.copy()
    group_df['Cluster'] = kmeans.fit_predict(X)
    final_clusters.append(group_df)

final_df = pd.concat(final_clusters).reset_index(drop=True)

final_df.head()

final_df.groupby(['Group', 'Cluster']).size()


cluster_summary = final_df.groupby(['Group', 'Cluster'])[per90_cols].mean().round(2)
display(cluster_summary)
```

```python
# Interpretation of these cluster groupings
```

```python
# cluster_summary = final_df.groupby(['Group', 'Cluster'])[per90_cols].mean().round(2)
import matplotlib.pyplot as plt
import numpy as np
# Loop through each Group (e.g. DEF, MID, ATT)

custom_names = {
    'Attacker': ['Elite Goalscorer', 'Reliable Goal-Volume Forward', 'Elite Playmaker', 'Defensive/Utility Attacker'],
    'Defender': ['Anchors & Pure CBs', 'Build-up Defenders', 'High-Progression Wing-Back', 'Defensive/Balanced Full-Back', 'Aggressive Ball-Carrying Defender'],
    'Midfielder': ['Anchor-Engine Hybrid', 'Midfield Controllers', 'Elite Advanced Playmakers', 'Utility/Secondary Goal Threat']
}

for group in cluster_summary.index.get_level_values('Group').unique():
    group_clusters = cluster_summary.loc[group]
    
    for cluster in group_clusters.index:
        stats = group_clusters.loc[cluster]
        labels = stats.index.tolist()
        values = stats.values.tolist()

        # Radar chart setup
        angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
        values += values[:1]
        angles += angles[:1]

        fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
        ax.plot(angles, values, linewidth=2)
        ax.fill(angles, values, alpha=0.25)
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(labels, fontsize=9)
        ax.set_yticklabels([])

        # Map cluster ID to custom name
        cluster_name = custom_names[group][cluster]
        ax.set_title(f'{group} - {cluster_name}', fontsize=13, fontweight='bold')

        plt.tight_layout()
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from adjustText import adjust_text 

for group_name, group_df in final_df.groupby('Group'):
    X = group_df[per90_cols].values

    # Run PCA
    pca = PCA(n_components=2)
    pca_coords = pca.fit_transform(X)

    group_df = group_df.copy()
    group_df['PCA1'] = pca_coords[:, 0]
    group_df['PCA2'] = pca_coords[:, 1]

    # Compute centroids in PCA space
    centroids = group_df.groupby('Cluster')[['PCA1', 'PCA2']].mean()

    plt.figure(figsize=(10, 7))
    all_texts = []

    for cluster_id, cluster_data in group_df.groupby('Cluster'):
        # Get custom cluster name
        cluster_label = custom_names[group_name][cluster_id]

        plt.scatter(
            cluster_data['PCA1'], cluster_data['PCA2'],
            label=cluster_label, alpha=0.6
        )

        # Centroid in PCA space
        centroid = centroids.loc[cluster_id].values.reshape(1, -1)

        # Distance from centroid
        distances = np.linalg.norm(cluster_data[['PCA1', 'PCA2']].values - centroid, axis=1)

        # Get top 3 closest & farthest players
        closest_idxs = distances.argsort()[:3]
        farthest_idxs = distances.argsort()[-3:]

        # Annotate closest players (bold black)
        for idx in closest_idxs:
            row = cluster_data.iloc[idx]
            txt = plt.text(
                row['PCA1'], row['PCA2'], row['Player'],
                fontsize=8, color='black', weight='bold'
            )
            all_texts.append(txt)

        # Annotate farthest players (italic red for emphasis)
        for idx in farthest_idxs:
            row = cluster_data.iloc[idx]
            txt = plt.text(
                row['PCA1'], row['PCA2'], row['Player'],
                fontsize=8, color='blue', style='italic', weight='bold'
            )
            all_texts.append(txt)

        # Mark centroid
        plt.scatter(*centroid.T, color='black', marker='X', s=80)

    # Adjust text to avoid overlap
    adjust_text(
        all_texts, 
        arrowprops=dict(arrowstyle='-', color='gray', lw=0.5)
    )

    plt.title(f'{group_name} Clusters in PCA Space', fontsize=14)
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
